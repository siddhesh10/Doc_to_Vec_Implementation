```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
from gensim.models import Word2Vec
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
%matplotlib inline
```

```python
df=pd.read_csv('500_questions.csv',encoding = "ISO-8859-1")
df.head(10)
```

```python
import re
import string
from nltk.corpus import stopwords

REPLACE_BY_SPACE_RE = re.compile('[/(){}\[\]\|@,;]')
BAD_SYMBOLS_RE = re.compile('[^0-9a-z #+_]')
STOPWORDS = set(stopwords.words('english'))

def text_prepare(row):
    text=row['question']
    # 1. Lowercase text
    text = text.lower() 
    # 2. Replace '/', '()', '{}', '[]', '|', '@', ',', ';' by space
    text = REPLACE_BY_SPACE_RE.sub(' ',text)
    # 3. Delete the non-alphanumerics excluding '#', '+', '_'
    text = BAD_SYMBOLS_RE.sub('',text)
    # 4. Deleting numeric values
    text = ''.join([i for i in text if not i.isdigit()])
    # 5. Deleting stop words
    text_words = text.split()
    text_words = [t for t in text_words if t not in STOPWORDS]
    # removing repeated words from each sentence.fo 
    text_words=set(text_words)
    text_words=list(text_words)
    text_words=' '.join(text_words)
    
    return text_words
df['question_updated'] = df.apply (lambda row: text_prepare(row),axis=1)
```

```python
df.head()
```

```python
list_of_questions=[]


for i in df['question_updated']:
    list_of_questions.append(i)

print(len(list_of_questions))
```

```python
from nltk.tokenize import word_tokenize
tagged_data = [TaggedDocument(words=word_tokenize(_d.lower()), tags=[str(i)]) for i, _d in enumerate(list_of_questions)]
```

```python
max_epochs = 100
vec_size = 20
alpha = 0.025

model = Doc2Vec(size=vec_size,alpha=alpha, min_alpha=0.00025,min_count=1,dm =1)
  
model.build_vocab(tagged_data)

for epoch in range(max_epochs):
    model.train(tagged_data,total_examples=model.corpus_count,epochs=model.iter)
    # decrease the learning rate
    model.alpha -= 0.0002
    # fix the learning rate, no decay
    model.min_alpha = model.alpha

model.save("d2v.model")
print("Model Saved")
```

```python
from gensim.models.doc2vec import Doc2Vec
model= Doc2Vec.load("d2v.model")
similar_doc = model.docvecs.most_similar('499')
print(similar_doc)
```

```python
list_of_similar_questions=[]
for i in range(500):
    small_list=[]
    k=str(i)
    similar_doc = model.docvecs.most_similar(k)
    
    for j in range(len(similar_doc)):
        small_list.append(int(similar_doc[j][0]))
    list_of_similar_questions.append(small_list)
```

```python
df_topic=pd.read_csv('topic_dataset.csv',encoding = "ISO-8859-1")
df_topic.head()
```

```python
#Searching questions topic wise if the word exist in 500 questions

from collections import defaultdict
d = []
my_count=0
my_count_no_check=0
my_count_check=0


m_c_n_c=[]
m_c_c=[]

for i in df_topic['topic']:
    x=[]
    c=0

    for j in df['question']:
        if i.lower() in j.lower():
            x.append(c)
        c=c+1
    d.append(x)

for i in range(len(d)):
    print("{0}--->{1}..................{2}".format(df_topic['topic'][i],d[i],my_count))
    my_count=my_count+1
    

print("\n\nNo need to check")
for i in range(len(d)):
    l=len(d[i])
    if l <=10:
        print("{0}--->{1}............{2}".format(df_topic['topic'][i],d[i],my_count_no_check))
        m_c_n_c.append(my_count_no_check)
    my_count_no_check=my_count_no_check+1

print("\n\nWe need to check")   
for i in range(len(d)):
    l=len(d[i])
    if l >10:
        print("{0}--->{1}...........{2}".format(df_topic['topic'][i],d[i],my_count_check))
        m_c_c.append(my_count_check)
    my_count_check=my_count_check+1
```

```python
# saving questions we don't need to check
#fcfs
#dml
#jdbc
#joints
#buffering
#applet
#ddl
#caching
#sheduling
n_t_c=[]
for i in m_c_n_c:
    for j in d[i]:
        #print(df_topic['topic'][i],df['question'][j])
        minor=[]
        minor.append(df['question'][j])
        
    n_t_c.append(minor)
```

```python
# we should check ones
import copy
from collections import Counter
final_list=[]
for i in m_c_c:
    temp_list=copy.deepcopy(d[i])
    #print(temp_list)
    for j in temp_list:
        temp_list=temp_list+list_of_similar_questions[j]
    count = Counter(temp_list)
    x=(count.most_common()[:10])
    final_list.append(x)
```

```python
to_check=[]
for i in final_list:
    q=[]
    for j in i:
        q.append(df['question'][j[0]])
    to_check.append(q)
        
```

```python
final_ans_list=n_t_c + to_check 
```

```python
df_ans = pd.DataFrame(final_ans_list)
```
